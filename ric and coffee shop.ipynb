### 6. Analyze Each Neighborhood

```python
# one hot encoding
onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
onehot['Neighborhoods'] = venues_df['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [onehot.columns[-1]] + list(onehot.columns[:-1])
onehot = onehot[fixed_columns]

print(onehot.shape)
onehot.head()
```

##### group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
grouped = onehot.groupby(["Neighborhoods"]).mean().reset_index()

print(grouped.shape)
grouped
```

```python
len(grouped[grouped["Coffee Shop"] > 0])
```

```python
CoffeeShop = grouped[["Neighborhoods","Coffee Shop"]]
```

```python
CoffeeShop
```

### 7. Cluster Neighborhoods

Run k-means to cluster the neighborhoods in region of Universities in Beijing into 3 clusters.

```python
# set number of clusters
kclusters = 5

clustering = CoffeeShop.drop(["Neighborhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
merged = CoffeeShop.copy()

# add clustering labels
merged["Cluster Labels"] = kmeans.labels_
```

```python
merged.rename(columns={"Neighborhoods": "Neighborhood"}, inplace=True)
merged
```

```python

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
merged = merged.join(df.set_index("Neighborhood"), on="Neighborhood")

print(merged.shape)
merged.head() 
```

```python
merged.head()
```

```python
# sort the results by Cluster Labels
print(merged.shape)
merged.sort_values(["Cluster Labels"], inplace=True)
merged
```

```python
import numpy as np 
# create map
map_clusters = folium.Map(location=[latitude, longitude])

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(merged['Latitude'], merged['Longitude'], merged['Neighborhood'], merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' - Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
